```python
def preprocess(optimalcube,velocitymap,dispersionmap,snrmap,noise_level=3):
    '''
    before plot the image, do
    preprocess to image, optimalcube
    should be an optimalcube with optimal
    image
    '''
    velocitymap[np.isnan(velocitymap)]=0
    dispersionmap[np.isnan(dispersionmap)]=0


    optimalimg_inter,x_inter,y_inter=Img_interpsmooth(optimalcube.optimal_img,
                                                     optimalcube.ra.to(u.arcsec).value,
                                                     optimalcube.dec.to(u.arcsec).value,[5,9])
    snrmap_inter,_,_=Img_interpsmooth(snrmap,
                                     optimalcube.ra.to(u.arcsec).value,
                                     optimalcube.dec.to(u.arcsec).value,[5,9])

    velocitymap_inter,_,_=Img_interpsmooth(velocitymap,
                                          optimalcube.ra.to(u.arcsec).value,
                                          optimalcube.dec.to(u.arcsec).value,[5,9])

    dispersionmap_inter,_,_=Img_interpsmooth(dispersionmap,
                                            optimalcube.ra.to(u.arcsec).value,
                                            optimalcube.dec.to(u.arcsec).value,[5,9])

    optimalimg_inter[snrmap_inter<noise_level]=999
    velocitymap_inter[snrmap_inter<noise_level]=np.nan
    dispersionmap_inter[snrmap_inter<noise_level]=np.nan
    

    
    return optimalimg_inter,velocitymap_inter,\
           dispersionmap_inter,snrmap_inter,x_inter,y_inter
```

```python
optimalimg,velocitymap,dispersionmap,snrmap,x,y=preprocess(lyoptimalcube,lyvelocitymap,
                                                          lydispersionmap,lysnrmap)
optimalimg[optimalimg==999]=np.power(10,-2.5)
optimalimg=optimalimg*1e3
noise=np.random.normal(0,.3,optimalimg.shape)
optimalimg+=noise

dirc='/Users/shiwuzhang/WS/ASTRO/MAMMOTH_KCWI/analysis results/newresults/ly/'
np.savetxt(dirc+'lyimg.out',optimalimg,delimiter=',')
np.savetxt(dirc+'lyvelo.out',velocitymap,delimiter=',')
np.savetxt(dirc+'lysnr.out',snrmap,delimiter=',')
np.savetxt(dirc+'lydisp.out',dispersionmap,delimiter=',')
np.savetxt(dirc+'lyx',x,delimiter=',')
np.savetxt(dirc+'lyy',y,delimiter=',')

# sources_coor=np.array([[40.052625, 220.3519792], [40.05221366, 220.3519858],
#                   [40.05297755, 220.3520401], [40.05222735, 220.3493004],
#                   [40.0534915, 220.3529473], [40.05224151, 220.3531833],
#                   [40.0535024, 220.349928],[40.05106072, 220.3497904]])*u.deg
# sources_name=['Source-B','','','G-6',
#               'G-1','G-3','G-5','G-4']
# refpoint=np.array([40.052625, 220.3519792])*u.deg
# sources_coor=(sources_coor-refpoint).to(u.arcsec).value
```

```python
optimalimg,velocitymap,dispersionmap,snrmap,x,y=preprocess(civoptimalcube,civvelocitymap,
                                                          civdispersionmap,civsnrmap,noise_level=2.)

optimalimg[optimalimg==999]=np.power(10,-2.8)
optimalimg=optimalimg*1e3
noise=np.random.normal(0,.03,optimalimg.shape)
optimalimg+=noise


sources_coor=np.array([[40.052625, 220.3519792], [40.05221366, 220.3519858],
                  [40.05297755, 220.3520401],[40.0534915, 220.3529473],
                  [40.05224151, 220.3531833]])*u.deg
refpoint=np.array([40.052625, 220.3519792])*u.deg
sources_name=['Source-B','','',
              'G-1','G-3']
sources_coor=(sources_coor-refpoint).to(u.arcsec).value
```

```python
IMG=[optimalimg,velocitymap,dispersionmap]
colornorm=[colors.SymLogNorm(linthresh=3, linscale=0.01),
           colors.SymLogNorm(linthresh=650, linscale=3),
           colors.SymLogNorm(linthresh=650, linscale=1)]
colormap=['Spectral_r','RdYlBu_r','RdYlBu_r']
fig,AX=plt.subplots(3,1,sharex=True,sharey=True)
plt.style.use('dark_background')
AX=AX.flatten()
AX[0].contour(x,y,snrmap,levels=[3,11,19,31,43])
for i in range(3):
    img=AX[i].contourf(x,y,IMG[i],cmap=colormap[i],levels=40,
                       norm=colornorm[i])
    AX[i].tick_params(axis='both', labelsize=15)
    cbar=fig.colorbar(img, ax=AX[i], pad=0.01)
    cbar.ax.tick_params(labelsize=15.)
#     AX[i].colorbar(img)
    AX[i].set_aspect('equal')
    for j in range(sources_coor.shape[0]):
        if j<=2:
            mak='*'
        else:
            mak='+'
        AX[i].scatter(sources_coor[j,0],sources_coor[j,1],s=180.,
                      linewidth=2.3, color='black',
                      marker=mak)
        AX[i].text(sources_coor[j,0]+.8,sources_coor[j,1]-.8,
                   sources_name[j],fontsize=15)
AX[0].set_title('Ly',color='red',fontsize=30)
AX[1].set_ylabel(r'$\Delta RA$(arcsec)',fontsize=20)
plt.subplots_adjust(wspace=0, hspace=0.05)
plt.rcParams['figure.figsize'] = [20, 20]
# plt.savefig('ly.png',dpi=300)
plt.show()
```

```python
lyslicecube=lynemissioncube.with_spectral_unit(u.km/u.s,velocity_convention='relativistic',
                                               rest_value=3.31*1215.67*u.AA)
lyslicecube.slice_generate(5)
```

```python
Slices=[]
Slices_loc=[]
MAX,MIN=[],[]
i=0
x,y=lyoptimalcube.ra.to(u.arcsec).value,lyoptimalcube.dec.to(u.arcsec).value
for slices in lyslicecube.slices[0]:
    MAX.append(np.max(slices))
    MIN.append(np.min(slices))
    slices[lysnrmap<3]=0
    img_inters,y_inters,x_inters=Img_interpsmooth(slices,x,y,[5,9])
    Slices.append(img_inters)
    Slices_loc.append(np.mean(np.array(lyslicecube.slices[1][i])))
    i+=1
```

```python
sources_coor=np.array([[40.052625, 220.3519792], [40.05221366, 220.3519858],
                  [40.05297755, 220.3520401], [40.05222735, 220.3493004],
                  [40.0534915, 220.3529473], [40.05224151, 220.3531833],
                  [40.0535024, 220.349928],[40.05106072, 220.3497904]])*u.deg
refpoint=np.array([40.052625, 220.3519792])*u.deg
sources_coor=(sources_coor-refpoint).to(u.arcsec).value
sources_coor
```

```python
MAX,MIN=np.array(MAX),np.array(MIN)
fig,AX=plt.subplots(4,4,sharex=True,sharey=True)
AX=AX.flatten()
for i in range(len(AX)-1):
    img=AX[i].contourf(y_inters,x_inters,Slices[i],cmap='Spectral_r',
                   vmax=np.max(MAX),vmin=np.min(MIN),levels=20)
    cbar=fig.colorbar(img, ax=AX[i], pad=0.01)
    circle=plt.Circle((0,0), 1, color='black', fill=False)
    AX[i].add_patch(circle)
    AX[i].set_title(str(int(Slices_loc[i]))+' km/s',fontsize=15)
    AX[i].tick_params(axis='both', labelsize=15)
    for j in range(sources_coor.shape[0]):
        if j<=2:
            mak='*'
        else:
            mak='+'
        AX[i].scatter(sources_coor[j,0],sources_coor[j,1],s=180.,
                      linewidth=1., color='black',
                      marker=mak)
    AX[i].set_aspect('equal')
fig.text(0.48, 0.1, r'$\Delta DEC$(arcsec)', ha='center', fontsize=25.)
fig.text(0.08, 0.5, r'$\Delta RA$(arcsec)', va='center', rotation='vertical', fontsize=25.)
plt.subplots_adjust(wspace=0.05, hspace=0.05)
fig.delaxes(AX[-1])
plt.rcParams['figure.figsize'] = [20, 20]
plt.savefig('lyslices.png')
```
