```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from spec_analysis import para_ly,para_heii,para_civ \
, out_ly, out_heii, out_civ \
, x_window_ly, y_window_ly, x_window_heii, y_window_heii \
, x_window_civ, y_window_civ \
, v_window_ly, sig_window_ly, v_window_heii, sig_window_heii \
, v_window_civ, sig_window_civ

para_ly=pd.DataFrame(para_ly)
para_heii=pd.DataFrame(para_heii)
para_civ=pd.DataFrame(para_civ)
```

```python
def img_map(img,x_window,y_window,fill_value,delta=10):
    
    img_new=img.copy()
    for i in range(x_window.size):
        img_new[x_window[i]-delta:x_window[i]+delta,
                y_window[i]-delta:y_window[i]+delta]=fill_value[i]
        
    return img_new
```

```python
arr_index=np.argsort(para_ly[['g1_center','g2_center','g3_center']],axis=1)
arr_index=np.array(arr_index)
v=np.array(para_ly[['g1_center','g2_center','g3_center']])
a=np.array(para_ly[['g1_amplitude','g2_amplitude','g3_amplitude']])
sig=np.array(para_ly[['g1_sigma','g2_sigma','g3_sigma']])
for i in range(arr_index.shape[0]):
    v[i]=v[i,arr_index[i]]
    a[i]=a[i,arr_index[i]]
    sig[i]=sig[i,arr_index[i]]
a=a/a.max(axis=1).reshape((29,1))
```

```python
cri=v[:,0]
cri=np.array(cri)
img=np.zeros((400,400))
img_new=img_map(img,x_window_ly,y_window_ly,cri)
plt.imshow(img_new,cmap='coolwarm',vmax=400,vmin=-400)
plt.colorbar()
plt.show()
```

```python
cri=v[:,2]
cri=np.array(cri)
img=np.zeros((400,400))
img_new=img_map(img,x_window_ly,y_window_ly,cri)
plt.imshow(img_new,cmap='coolwarm',vmax=1000,vmin=-1000)
plt.colorbar()
plt.show()
```
